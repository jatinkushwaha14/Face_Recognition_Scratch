```python
from tensorflow import keras
import os
import numpy as np
import cv2 as cv
model = keras.models.load_model('/Users/jatinkushwaha/My-Files/PYTHON MAIN/Test/FaceRecogScratch/TrainedFaceRecognition.h5')
```

```python
def resize_bbox(original_size, target_size, bbox):
    original_width, original_height = original_size
    target_width, target_height = target_size

    x_scale = target_width / original_width
    y_scale = target_height / original_height

    x1, y1, x2, y2 = bbox
    x1 = int(x1 * x_scale)
    y1 = int(y1 * y_scale)
    x2 = int(x2 * x_scale)
    y2 = int(y2 * y_scale)

    return [x1, y1, x2, y2]
```

```python
cap = cv.VideoCapture(0)
while(True):
    ret, frame = cap.read()
    if not ret:
        print("Failed to capture image")
        break
    frame_resized = cv.resize(frame, (244, 244))
    frame_normalized = frame_resized.astype('float32') / 255.0
    image_expanded = np.expand_dims(frame_normalized, axis=0)

    predictions = model.predict(image_expanded)
    label_predictions = predictions[0]
    bbox_predictions = predictions[1]

    pred_x1 = int(bbox_predictions[0][0] * frame.shape[1])
    pred_y1 = int(bbox_predictions[0][1] * frame.shape[0])
    pred_x2 = int(bbox_predictions[0][2] * frame.shape[1])
    pred_y2 = int(bbox_predictions[0][3] * frame.shape[0])

    if label_predictions[0] > 0.5:
        cv.rectangle(frame, (pred_x1, pred_y1), (pred_x2, pred_y2), (255, 0, 0), 2)
        
    cv.imshow('Prediction', frame)

    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
